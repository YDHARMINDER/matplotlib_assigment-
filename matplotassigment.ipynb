ans3. subplots method provides a way to plot multiple plots on a single figure. Given the number of rows and columns , it returns a tuple ( fig , ax ), giving a single figure fig with an array of axes ax .
The subplot() function takes three arguments that describes the layout of the figure. The layout is organized in rows and columns, which are represented by the first and second argument. The third argument represents the index of the current plot. 


```python
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
y1 = np.array([50, 20, 40, 20, 60, 70])
y2 = np.array([10, 20, 30, 40, 50, 60])
y3 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig,a =  plt.subplots(2,2)
a[0][0].plot(x,y)
a[0][0].set_title('x,y')
a[0][1].plot(x,y1)
a[0][1].set_title('x, y1')
a[1][0].plot(x,y2)
a[1][0].set_title('x,y2')
a[1][1].plot(x,y3)
a[1][1].set_title('x, y3')
plt.show()
```

ans4. Bar plots are a type of data visualization used to represent data in the form of rectangular bars. The height of each bar represents the value of a data point, and the width of each bar represents the category of the data. The Matplotlib library in Python is widely used to create bar plots.


The bars can be used to plotted vertically or horizontally. A bar graph shows comparisons among discrete categories.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company, profit)
plt.show()
```
